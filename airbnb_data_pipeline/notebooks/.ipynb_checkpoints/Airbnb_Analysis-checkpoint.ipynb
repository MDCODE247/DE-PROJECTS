```python
# -----------------------------
# Step 0: Setup - Create folders
# -----------------------------
import os
import pandas as pd
import requests
from io import BytesIO

# Create project folders automatically
os.makedirs("data", exist_ok=True)
os.makedirs("output", exist_ok=True)
os.makedirs("notebooks", exist_ok=True)
print("✅ Project folders ready!")

# -----------------------------
# Step 1: Download dataset
# -----------------------------
csv_url = "http://data.insideairbnb.com/united-states/ny/new-york-city/2025-11-01/data/listings.csv.gz"

response = requests.get(csv_url)
if response.status_code == 200:
    print("✅ CSV downloaded successfully!")
else:
    raise Exception("❌ Could not download CSV")

# Load CSV into Pandas directly
df = pd.read_csv(BytesIO(response.content), compression='gzip')
print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")

# Optional: save raw CSV locally
df.to_csv("data/listings.csv", index=False)
print("✅ Raw CSV saved to data/listings.csv")

# -----------------------------
# Step 2: Data Cleaning
# -----------------------------
# Convert price to float (fixed raw string)
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)

# Convert last_review column to datetime
df['last_review'] = pd.to_datetime(df['last_review'])

```

```python
# -----------------------------
# Step 2: Data Cleaning
# -----------------------------

# Convert price to float (raw string fix)
df['price'] = df['price'].replace(r'[\$,]', '', regex=True).astype(float)

# Convert last_review to datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')  # coerce invalid dates to NaT

# Handle missing values
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['host_name'] = df['host_name'].fillna("Unknown")

# Drop rows with missing 'neighbourhood_group' if it exists
if 'neighbourhood_group' in df.columns:
    df = df.dropna(subset=['neighbourhood_group'])
else:
    print("⚠️ 'neighbourhood_group' column not found. Skipping this step.")

# Remove listings with zero price or minimum nights
df = df[(df['price'] > 0) & (df['minimum_nights'] > 0)]

# Save cleaned dataset
df.to_csv("output/listings_cleaned.csv", index=False)
print("✅ Cleaned dataset saved to output/listings_cleaned.csv")

# Verify cleaned data
df.info()

```

```python
# -----------------------------
# Step 3: Data Enrichment
# -----------------------------

# Make a safe copy of the cleaned DataFrame
df = df.copy()

# 1️⃣ Price per booking
df['price_per_booking'] = df['price'] * df['minimum_nights']

# 2️⃣ Availability categories
def availability_category(x):
    if x > 300:
        return 'Full-time'
    elif x >= 100:
        return 'Part-time'
    else:
        return 'Rare'

df['availability_type'] = df['availability_365'].apply(availability_category)

# Optional: view first few rows to verify
df[['price', 'minimum_nights', 'price_per_booking', 'availability_365', 'availability_type']].head()
```

```python
# -----------------------------
# Step 4: Data Analysis
# -----------------------------

# 4.1 Top 10 most expensive neighborhoods (by average price)
top10_expensive = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)
print("Top 10 most expensive neighborhoods:\n", top10_expensive)

# 4.2 Average availability and price by room type
avg_room_stats = df.groupby('room_type')[['price', 'availability_365']].mean()
print("\nAverage price & availability by room type:\n", avg_room_stats)

# 4.3 Host with the most listings
top_host = df['host_name'].value_counts().head(1)
print("\nHost with the most listings:\n", top_host)

# 4.4 Average price by borough/district
# Use 'neighbourhood_group_cleansed' since 'neighbourhood_group' is missing
avg_price_borough = df.groupby('neighbourhood_group_cleansed')['price'].mean()
print("\nAverage price by borough/district:\n", avg_price_borough)

# 4.5 Listings never reviewed
never_reviewed = df[df['number_of_reviews'] == 0].shape[0]
print(f"\nListings never reviewed: {never_reviewed}")

# -----------------------------
# Step 5: Save analysis results
# -----------------------------
top10_expensive.to_csv("output/top10_expensive_neighbourhoods.csv")
avg_room_stats.to_csv("output/avg_room_stats.csv")
avg_price_borough.to_csv("output/avg_price_by_borough.csv")
print("\n✅ Analysis results saved in output/ folder")
```

```python
# -----------------------------
# Step 6: Visualizations
# -----------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# 6.1 Top 10 most expensive neighborhoods
plt.figure(figsize=(10,6))
sns.barplot(x=top10_expensive.values, y=top10_expensive.index, palette="viridis")
plt.title("Top 10 Most Expensive Neighborhoods (Average Price)")
plt.xlabel("Average Price ($)")
plt.ylabel("Neighborhood")
plt.tight_layout()
plt.show()

# 6.2 Average price by room type
plt.figure(figsize=(8,5))
sns.barplot(x=avg_room_stats.index, y=avg_room_stats['price'], palette="coolwarm")
plt.title("Average Price by Room Type")
plt.ylabel("Average Price ($)")
plt.xlabel("Room Type")
plt.tight_layout()
plt.show()

# 6.3 Average availability by room type
plt.figure(figsize=(8,5))
sns.barplot(x=avg_room_stats.index, y=avg_room_stats['availability_365'], palette="magma")
plt.title("Average Availability by Room Type")
plt.ylabel("Average Availability (days/year)")
plt.xlabel("Room Type")
plt.tight_layout()
plt.show()

# 6.4 Average price by borough/district
plt.figure(figsize=(8,5))
sns.barplot(x=avg_price_borough.index, y=avg_price_borough.values, palette="Set2")
plt.title("Average Price by Borough/District")
plt.ylabel("Average Price ($)")
plt.xlabel("Borough/District")
plt.tight_layout()
plt.show()

# 6.5 Listings never reviewed
plt.figure(figsize=(5,5))
sns.countplot(x=df['number_of_reviews'].apply(lambda x: 'Never Reviewed' if x==0 else 'Reviewed'), palette="pastel")
plt.title("Listings Never Reviewed vs Reviewed")
plt.ylabel("Count of Listings")
plt.xlabel("")
plt.tight_layout()
plt.show()
```
